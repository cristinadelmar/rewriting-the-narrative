```python
all_results = []

for i in range(15):
    page_url = base_url + "&start=" + str(i * 10)
    page_html = requests.get(page_url).text
    page_soup = BeautifulSoup(page_html, "html.parser")
    article_els = page_soup.select("article.teaser")

    for article in article_els:
        title_element = article.select("h2 span")
        title = title_element[0].text.strip() if title_element else "No title available"

        preview_element = article.select("div.teaser-body")
        preview = preview_element[0].text.strip()

        date_element_modified = article.select(".teaser-article-modifieddate")
        date_element_pubdate = article.select(".teaser-article-pubdate")

        if date_element_modified:
            article_date = date_element_modified[0].text.replace("Updated: ", "")
        elif date_element_pubdate:
            article_date = date_element_pubdate[0].text.replace("Published: ", "")
        else:
            article_date = "No date available"
        
        link = article.select("a")[0]["href"].strip()
        
        all_results.append({'Title': title, 'Preview': preview, 'Date': article_date, 'Link': link})
```

```python
df_all_results = pd.DataFrame(all_results)
df_all_results.head(25)
```

```python
# Convert the 'Date' column to datetime objects
df_all_results['Date'] = pd.to_datetime(df_all_results['Date'], errors='coerce')

df_all_results.head(25)
```

```python
# Remove the time
df_all_results['Date'] = df_all_results['Date'].dt.date

df_all_results.head(25)
```

### Ahora quiero sacar todos los que tienen la palabra feminicidio o transfeminicidio en el título o en el preview por año



```python
# Filtering rows with the word "femicide" in the title or preview
filtered_df = df_all_results[df_all_results['Title'].str.contains('feminicidio', case=False) | df_all_results['Preview'].str.contains('feminicidio', case=False)]

# Step 2: Convert the 'date' column to pandas datetime format
filtered_df['Date'] = pd.to_datetime(filtered_df['Date'])

# Step 3: Create new columns for month and year
#filtered_df['Month'] = filtered_df['Date'].dt.month
filtered_df['Year'] = filtered_df['Date'].dt.year

# Step 4: Group by month and year and perform any desired aggregation (e.g., count the occurrences)
grouped_df = filtered_df.groupby(['Year']).size().reset_index(name='Count')

# Resulting 'grouped_df' will have columns 'year', 'month', and 'count' indicating the occurrences of "femicide" in each month and year.

grouped_df
```

```python
#Articles with 'feminicidio' in title

all_feminicidio_titles = []

for i in range(15):
    page_url = base_url + "&start=" + str(i * 10)
    page_html = requests.get(page_url).text
    page_soup = BeautifulSoup(page_html, "html.parser")
    article_els = page_soup.select("article.teaser")

    for article in article_els:
        title_element = article.select("h2 span")
        title = title_element[0].text.strip() if title_element else "No title available"

        # Check if the title contains the word "feminicidio"
        if "feminicidio" in title.lower():
            preview_element = article.select("div.teaser-body")
            preview = preview_element[0].text.strip() if preview_element else "No preview available"

            l_update_element = article.select(".teaser-article-modifieddate")
            l_update = l_update_element[0].text if l_update_element else "No date available"

            link = article.select("a")[0]["href"].strip()

            all_feminicidio_titles.append({'Title': title, 'Preview': preview, 'Date': l_update, 'Link': link})

all_feminicidio_titles[:3]
```
