```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

import time
```

```python
# Define the base URL of the news website
base_url = 'https://www.elvocero.com/search/?l=25&s=start_time&sd=desc&f=html&t=article%2Cvideo%2Cyoutube%2Ccollection&app%5B0%5D=editorial&nsa=eedition&q=feminicidio&o=0'


# Create an empty dataframe to store the scraped data

df = pd.DataFrame(columns=['Date', 'Title', 'Preview', 'Link'])
```

```python
for i in range(19):
    print(base_url + "&o=" + str(i * 25))
```

```python
all_results = []

for i in range(19):
    page_url = base_url + "&o=" + str(i * 25)
    page_html = requests.get(page_url).text
    page_soup = BeautifulSoup(page_html, "html.parser")
    article_els = page_soup.select("div.card-body")

    for article in article_els:
        title_element = article.select("h3 a")
        title = title_element[0].text.strip() if title_element else "No title available"
        print(title)

        preview_element = article.select("p.tnt-summary")
        if preview_element:
            preview = preview_element[0].text.strip()
            print(preview)
        else:
            preview = "No preview available"
            print(preview)

        date_element = article.select("li time")
        if date_element:
            date = date_element[0].text.strip()
            print(date)
        else:
            date = "No date available"
            print(date)

        link_element = article.select("h3 a")
        link = link_element[0]['href'] if link_element else "No link available"
        print(link)

        all_results.append({'Title': title, 'Preview': preview, 'Date': date, 'Link': link})
```

```python
df_all_results = pd.DataFrame(all_results)
df_all_results.head(25)
```

```python
df_all_results = df_all_results.drop(df_all_results[df_all_results['Title'] == 'No title available'].index)
df_all_results = df_all_results.reset_index(drop=True)
df_all_results.head(25)
```

```python
# Convert the 'Date' column to datetime objects
df_all_results['Date'] = pd.to_datetime(df_all_results['Date'], errors='coerce')

df_all_results.head(25)
```

```python
#Replace the most recent article's date to today's date:

df_all_results['Date'] = df_all_results['Date'].fillna(pd.to_datetime('2023-07-21'))
df_all_results.head(3)
```

```python
# Filtering rows with the word "femicide" in the title or preview
filtered_df = df_all_results[df_all_results['Title'].str.contains('feminicidio', case=False) | df_all_results['Preview'].str.contains('feminicidio', case=False)]

# Step 2: Convert the 'date' column to pandas datetime format
filtered_df['Date'] = pd.to_datetime(filtered_df['Date'])

# Step 3: Create new columns for month and year
#filtered_df['Month'] = filtered_df['Date'].dt.month
filtered_df['Year'] = filtered_df['Date'].dt.year

# Step 4: Group by month and year and perform any desired aggregation (e.g., count the occurrences)
grouped_df = filtered_df.groupby(['Year']).size().reset_index(name='Count')

# Resulting 'grouped_df' will have columns 'year', 'month', and 'count' indicating the occurrences of "femicide" in each month and year.

grouped_df
```

```python
grouped_df.to_csv('output_ElVocero.csv', index=False)
```
